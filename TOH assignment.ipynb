```python
data3=pd.read_csv("BuyerRatio.csv")
```

```python
data3
```

```python
df_table=data3.iloc[:,1:6]
df_table
```

```python
obs=df_table.values
obs1=np.array(obs)
```

```python
exp=chi2_contingency(obs)
exp1=exp.expected_freq
```

```python
chi_square=sum([(o-e)**2/e for o,e in zip(df_table.values,exp1)])
chi_square_statestic=chi_square.sum()
chi_square_statestic
```

```python
row=len(data3.iloc[0:2,0])
col=len(data3.iloc[0,0:4])
degree_of_freedom=(row-1)*(col-1)
```

```python
critical_value=chi2.ppf(0.95,3)
```

```python
if chi_square_statestic>=critical_value:
  print("Reject H0")
else:
  print("Accept H0")
```

```python
pvalue=chi2.sf(chi_square_statestic,3)
pvalue
```

```python
if pvalue>=0.05:
  print("Accept H0")
else:
  print("Areject H0")
```
